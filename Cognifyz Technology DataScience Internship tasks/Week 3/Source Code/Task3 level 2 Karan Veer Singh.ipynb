```python
plt.figure(figsize=(10,4))
plt.hist(df['Restaurant Name'].str.len(), bins=20, color='skyblue', edgecolor='black')
plt.title('Name Length Before FE')
plt.xlabel('Length')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10,4))
plt.hist(df['Address'].str.len(), bins=20, color='salmon', edgecolor='black')
plt.title('Address Length Before FE')
plt.xlabel('Length')
plt.ylabel('Count')
plt.show()
```

```python
df['name_len']=df['Restaurant Name'].str.len()
df['addr_len']=df['Address'].str.len()
df['tb_bin']=df['Has Table booking'].map({'Yes':1,'No':0})
df['od_bin']=df['Has Online delivery'].map({'Yes':1,'No':0})
```

```python
plt.figure(figsize=(10,4))
plt.hist(df['name_len'], bins=20, color='skyblue', edgecolor='black')
plt.title('Name Length After FE')
plt.xlabel('Length')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10,4))
plt.hist(df['addr_len'], bins=20, color='salmon', edgecolor='black')
plt.title('Address Length After FE')
plt.xlabel('Length')
plt.ylabel('Count')
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,4))

axes[0].bar([0,1], df['tb_bin'].value_counts().sort_index(), color=['lightgreen','lightcoral'], edgecolor='black')
axes[0].set_title('Table Booking Binary')
axes[0].set_xticks([0,1])
axes[0].set_xticklabels(['No','Yes'])
axes[0].set_ylabel('Count')

axes[1].bar([0,1], df['od_bin'].value_counts().sort_index(), color=['lightblue','orange'], edgecolor='black')
axes[1].set_title('Online Delivery Binary')
axes[1].set_xticks([0,1])
axes[1].set_xticklabels(['No','Yes'])
plt.tight_layout()
plt.show()
```
