```python
#Karan Veer Singh
#Internship Task 3
#Cognifyz Tech 

import pandas as pd,numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('Dataset .csv')
print(df.shape)
df.head()
```

```python
city_ct = df['City'].value_counts()
country_ct = df['Country Code'].value_counts()
print('Cities count:')
print(city_ct.head())
print()
print('Country code counts:')
print(country_ct.head())
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Longitude'],df['Latitude'],alpha=0.4,c=df['Aggregate rating'],cmap='viridis',s=15)
plt.colorbar(label='Aggregate rating')
plt.title('Restaurant locations colored by rating')
plt.xlabel('Longitude');plt.ylabel('Latitude')
plt.show()
```

```python
city_rating=df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,4))
plt.barh(city_rating.index,city_rating.values,color='darkcyan')
plt.title('Top 10 Cities by Avg Rating')
plt.xlabel('Avg Rating')
plt.tight_layout()
plt.show()
```

```python
country_rating=df.groupby('Country Code')['Aggregate rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,4))
country_rating.plot(kind='bar',color='cornflowerblue')
plt.title('Avg Rating by Country Code')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
print('Correlation longitude vs rating:',df['Longitude'].corr(df['Aggregate rating']))
print('Correlation latitude vs rating:',df['Latitude'].corr(df['Aggregate rating']))
print('Done geospatial analysis for',len(df),'restaurants')
```
